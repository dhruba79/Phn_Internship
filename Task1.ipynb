```python
import pandas as pd
```

```python
# 3

# Read CSV file and create a data frame
csv_file = pd.read_csv('C:/Users/DHRUBA/Desktop/ML/internship/a1.csv')

# Read Excel file and create a data frame
excel_file = pd.read_excel('C:/Users/DHRUBA/Desktop/ML/internship/A2.xlsx')

# Print the data frames
print(csv_file)
print(excel_file)
```

```python
# 4
```

```python
df = pd.merge(csv_file, excel_file, on='Company Name', how='inner')
```

```python
print(df)
```

```python
# 5
```

```python
df.isnull().sum()
```

```python
df = df.fillna(df.mean())

df
```

```python
# 6
```

```python
df['GST']=['.18','.12','.15','.17']
df
```

```python
# iterate over the merged data frame and add GST value to On road pricing
for index, row in df.iterrows():
    On_Road_Pricing = row['On Road Pricing']
    GST = row['GST']
    On_Road_Pricing += On_Road_Pricing + GST
    df.at[index, 'On Road Pricing'] = On_Road_Pricing
```
