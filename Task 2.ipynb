```python
#4
# calculate statistics aggregates using .agg() method
agg_dict = {'Age': ['min', 'max', 'mean', 'median', 'std'],
            'Fare': ['min', 'max', 'mean', 'median', 'std']}

stats = df.agg(agg_dict)

print(stats)
```

```python
#5
# group by 'Sex' and 'Cabin', then calculate mean fare
mean_fares = df.groupby(['Sex', 'Cabin']).mean()['Fare']

print(mean_fares)
```
